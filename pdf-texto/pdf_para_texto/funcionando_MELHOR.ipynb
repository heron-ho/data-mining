```python
from PyPDF2 import PdfReader
from sklearn.feature_extraction.text import CountVectorizer
import re
```

```python
# Obtendo o texto do pdf
reader = PdfReader('petro.pdf')
number_of_pages = len(reader.pages)
page = reader.pages[0]
text = page.extract_text()

texto = ''
for pagina in range(number_of_pages):
    texto += reader.pages[pagina].extract_text()
texto
```

```python
# Pré processando o texto
# Função para pré-processamento de texto
def preprocessamento(texto):
    # Transforma o texto em minúsculas
    texto = texto.lower()
    # Remove pontuações e caracteres especiais
    texto = re.sub(r'[^\w\s]', '', texto)
    return texto

# Pré processamento de texto
texto_preprocessado = preprocessamento(texto)
texto_preprocessado
```

```python
# Tokenização do texto
vectorizer = CountVectorizer(ngram_range=(1,2))
X = vectorizer.fit_transform([texto_preprocessado])
```

```python
# lista de palavras para conferir frequência

"""
Manufatura digital
Automação avançada
Produção inteligente
Fábrica inteligente
Logística digital
Internet industrial
Robótica avançada
Manutenção preditiva
Realidade aumentada
Gêmeos digitais
Impressão 3D
Análise de dados
Simulação industrial
Segurança cibernética
Fabricação aditiva
Plataformas digitais
Inovação tecnológica
Manufatura avançada
Monitoramento remoto
Inteligência Artificial
"""
```

```python
# Obtenção do índice do termo "inteligência artificial" no vetor de recursos
indice_inteligencia_artificial = vectorizer.vocabulary_.get('inteligência artificial')

# Se o termo não estiver presente, o índice será None
if indice_inteligencia_artificial is not None:
    # Recuperação da frequência do termo
    frequencia_inteligencia_artificial = X[0, indice_inteligencia_artificial]
    print("Frequência de 'inteligência artificial' no texto:", frequencia_inteligencia_artificial)
else:
    print("O termo 'inteligência artificial' não está presente no texto.")
```

```python
# Obtenção do índice do termo "sustentabilidade" no vetor de recursos
indice_sustentabilidade = vectorizer.vocabulary_.get('sustentabilidade')

# Se o termo não estiver presente, o índice será None
if indice_sustentabilidade is not None:
    frequencia_sustentabilidade = X[0, indice_sustentabilidade]
    print("Frequência de 'sustentabilidade' no texto:", frequencia_sustentabilidade)
else:
    print("O termo 'sustentabilidade' não está presente no texto.")
```
